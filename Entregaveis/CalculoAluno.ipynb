```python
'''
Created on 6 de mai. de 2024

@author: nelso
'''


class Aluno:

    def _init(self):
        self.nome = "";
        self.nota_av1 = 0.0;
        self.nota_av2 = 0.0;
        self.nota_av3 = 0.0;        
        self.media = 0.0; 
        self.status = False;
        
    def calcular_media(self):
        self.media = round((self.nota_av1 + self.nota_av2 + self.nota_av3)/3,2)
```

```python
'''
Created on 6 de mai. de 2024

@author: nelso
'''
from manipulacao.Aluno import Aluno


def gravar_turma():
    quantidade_alunos = int(input("Favor informar a quantidade de alunos na turma: "))
    cont = 1
    turma = []
    while(cont <= quantidade_alunos):
        aluno = Aluno()
        aluno.nome = input("Favor informar o nome do aluno " + str(cont)  + ": ")
        aluno.nota_av1 = float(input("Favor informar a primeira nota do aluno " + str(cont) + ": "))
        aluno.nota_av2 = float(input("Favor informar a segunda nota do aluno " + str(cont) + ": "))
        aluno.nota_av3 = float(input("Favor informar a terceira nota do aluno " + str(cont) + ": "))
        turma.append(aluno)
        cont += 1
        
    arquivo = open("Notas_Alunos.csv", "w")
    for a in turma:
        arquivo.write(a.nome + ";" + str(a.nota_av1) + ";" + str(a.nota_av2) + ";" + str(a.nota_av3) + "\n")    
        
    arquivo.close()
    input(" *** Cadastro realizado com sucesso ! ***")
    
def le_arquivo():
    arquivo = open("Notas_Alunos.csv", "r")
    tamanho = arquivo.readlines()
    cont = 0
    turma = []
    while(cont < len(tamanho)):
        linha = tamanho[cont]
        dados = linha.split(";")
        aluno = Aluno()
        aluno.nome = dados[0]
        aluno.nota_av1 = float(dados[1])
        aluno.nota_av2 = float(dados[2])
        aluno.nota_av3 = float(dados[3])
        aluno.calcular_media()
        
        if(aluno.media >= 6):
            aluno.status = True
        else:
            aluno.status = False
            
        turma.append(aluno)
        cont += 1
    
    arquivo_aprovado = open("Notas_Alunos_Aprovados.csv", "w")
    arquivo_reprovado = open("Notas_Alunos_Reprovados.csv", "w")
    
    for a in turma:
        if(a.status):
            arquivo_aprovado.write(a.nome + ";" + str(a.media) + "\n") 
        else:       
            arquivo_reprovado.write(a.nome + ";" + str(a.media) + "\n") 

    arquivo.close()
    arquivo_aprovado.close()
    arquivo_reprovado.close()
    input(" *** Cadastro realizado com sucesso ! ***")
    
    
le_arquivo()
```
