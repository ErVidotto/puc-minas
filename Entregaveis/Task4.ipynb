```python
class Automovel:
  def __init__(self):
    self.cor = None
    self.marca = None
    self.modelo = None
    self.ano_fabricacao = None
    self.numero_portas = None
    self.chassi = None
    self.placa = None


  def gravar(self):
     arquivo = open("Automovel.txt", "w")
     arquivo.write("Marca ...: " + self.marca + " \n")
     arquivo.write("Modelo ..: " + self.modelo + " \n")
     arquivo.write("Cor .....: " + self.cor  + " \n")
     arquivo.write("Placa ...: " + self.placa + " \n")
     arquivo.write("Chassi ..: " + self.chassi + " \n")
     arquivo.write("Nº Portas: " + str(self.numero_portas)+ " \n")
     arquivo.write("Ano Fabr.: " + str(self.ano_fabricacao)+ " \n")
     arquivo.close()

  def le_arquivo(self):
      arquivo = open("/content/Automovel.txt", "r", encoding="utf-8")
      conteudo = " "
      while(conteudo != ""):
        conteudo = arquivo.readline()
        print(conteudo)


  def imprimir(self):
     print("Marca ...: " + self.marca )
     print("Modelo ..: " + self.modelo )
     print("Cor .....: " + self.cor  )
     print("Placa ...: " + self.placa )
     print("Chassi ..: " + self.chassi )
     print("Nº Portas: " + str(self.numero_portas))
     print("Ano Fabr.: " + str(self.ano_fabricacao))

  def ligar(self):
    print("Automovel ligando ...")

  def desligar(self):
    print("Automovel desligando ...")

  def frear(self):
    print("Automovel freando ...")

  def acelerar(self):
    print("Automovel acelerando ...")

  def travar_portas(self):
    print("Automovel travando as portas ...")

for x in range(2):
  a = Automovel()
  a.marca = input("Digite a Marca do automovel: ")
  a.modelo = input("Digite a Modelo do automovel: ")
  a.cor = input("Digite a Cor do automovel: ")
  a.placa = input("Digite a Placa do automovel: ")
  a.chassi = input("Digite o chassi do automovel: ")
  a.numero_portas = int(input("Digite o número de portas automovel: "))
  a.ano_fabricacao = int(input("Digite o ano de fabricação do automovel: "))
  a.imprimir(a)
  a.ligar()
  a.travar_portas()
  a.acelerar()
  a.frear()
  a.desligar()
  a.gravar()
  a.le_arquivo()
```
